## F Strings in python
They make really easy to mix strings and different data types. The most convenient way of incorporating things that have different data types. Let's say instead of having to convert all and use a whole bunch of plus signs and then convert everything into a string, we can use use something in Python known as an F string.

F string allows us to do is essentially adding just the character f in front of the string, and you can start adding various values into this string. We can do that by putting that variable inside a set of curly braces, and now it does all of the converting

You can see below that our entire string now prints out your score is 0,

your height is 1.8 you are winning is True.

So all of these different data types got combined into a string by using an F in

front of the string and then using these curly braces to place our variables

into this string. By using f strings,

you cut down on a lot of the manual labor of inserting different data types into

a string.

```python
score = 0
height = 1.8
isWinning = True
#f-string
print(f"your score is {score}, your height is {height}, you are winning is {isWinning}")
```
